```python
data = pd.read_csv('car.data.txt', names=['Buying', 'Maint','doors', 'persons','lug-boot','safety','target'])
data.head()
```

```python
def Trans_Buying(x):
    
    if x== 'vhigh': 
        return 0
    if x=='high':
        return 1
    if x== 'med':
        return 2
    if x == 'low':
        return 3 
data['Buying']=data['Buying'].apply(Trans_Buying)
#print(data['Buying'])

def Trans_Maint(x):
    
    if x== 'vhigh': 
        return 0
    if x=='high':
        return 1
    if x== 'med':
        return 2
    if x == 'low':
        return 3 
data['Maint']=data['Maint'].apply(Trans_Maint)
#print(data['Maint'])

def Trans_doors(x):
    
    if x== '2': 
        return 0
    if x=='3':
        return 1
    if x== '4':
        return 2
    if x == '5more':
        return 3 
data['doors']=data['doors'].apply(Trans_doors)
#print(data['doors'])

def Trans_persons(x):
    
    if x== '2': 
        return 0
    if x=='4':
        return 1
    if x== 'more':
        return 2
   
data['persons']=data['persons'].apply(Trans_persons)
#print(data['persons'])

def Trans_lug_boot(x):
    
    if x== 'small': 
        return 0
    if x=='med':
        return 1
    if x== 'big':
        return 2
   
data['lug-boot']=data['lug-boot'].apply(Trans_lug_boot)
#print(data['lug-boot'])

def Trans_safety(x):
    
    if x== 'low': 
        return 0
    if x=='med':
        return 1
    if x== 'high':
        return 2
   
data['safety']=data['safety'].apply(Trans_safety)
#print(data['safety'])

def Trans_target(x):
    
    if x == 'unacc':
        return 0
    if x == 'acc':
        return 1
    if x == 'vgood':
        return 2
    if x == 'good':
        return 3

data['target'] = data['target'].apply(Trans_target)

data = data.fillna(0) #getting read from the nan values and replace them with zero
```

```python
from sklearn.datasets import load_iris
from sklearn import tree

columns = ['Buying', 'Maint','doors', 'persons','lug-boot','safety']
classNames = ['unacc', 'acc', 'vgood', 'good']

labels = data["target"].values
features = data[list(columns)].values

clf = tree.DecisionTreeClassifier(criterion = 'entropy')
clf = clf.fit(features, labels)
```

```python
import graphviz

dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=columns,  
                         class_names=classNames,  
                         filled=True, rounded=True,  
                         special_characters=True)

graph = graphviz.Source(dot_data)  
graph 
```
